# 1. Assemblage du système
___

## Laplace 1d

Shéma différence finies sur l'intervalle $[0,1]$ : $\dfrac{-u_{i+1}+2u_i-u_{i-1}}{h^2}$ la matrice est tridiagonale

```julia
function Laplace1d(n)
    h=1./(n+1)
    d1=[-1/h^2 for i=1:n-1];
    d2=[2/h^2 for i=1:n];
    d3=[-1/h^2 for i=1:n-1];
    spdiagm((d1,d2,d3),(-1,0,1));   
end
```

## Laplace 2d

Shéma différence finies sur le domaine $[0,1]^2$ : $\dfrac{-u_{i+n}-u_{i+1}+4u_i-u_{i-1}-u_{i-n}}{h^2}$ la matrice est pentadiagonale on utilise la multiplication de Kronecker

```julia
function Laplace2d(n,m)
    kron(speye(m),Laplace1d(n))+kron(Laplace1d(m),speye(n))
end
```

## Laplace 3d

Shéma différence finies sur le domaine $[0,1]^3$ : $\dfrac{-u_{i+n^2}-u_{i+n}-u_{i+1}+6u_i-u_{i-1}-u_{i-n}-u_{i-n^2}}{h^2}$ la matrice est heptadiagonale on utilise également la multiplication de Kronecker

```julia
function Laplace3d(n,m,p)
    kron(speye(p),Laplace2d(n,m))+kron(Laplace1d(p),speye(n*m))
end
```

## Test de performance d'assemblage : (matrices de taille max 10 000 000)

Le temps d'assemblage n'est pas forcément optimum, en effet le choix de la fonction `kron()` n'est pas forcément le meilleur... Néanmoins regardons le temps d'assemblage des matrices qui doit être linéaire en fonction de la taille de la diagonale de la matrice.

```julia
npt=12
println("1d... ");
N1d=iround(linspace(100,1e7,npt));
T1d=zeros(length(N1d));
for i=1:length(N1d) tic();Laplace1d(N1d[i]);T1d[i]=toq();end;
println("      completed!");

println("2d... ");
N2d=iround(linspace(10,10^3.5001,npt));
T2d=zeros(length(N2d));
for i=1:length(N2d) tic();Laplace2d(N2d[i],N2d[i]);T2d[i]=toq();end;
println("      completed!");

println("3d...");
N3d=iround(linspace(5,10^(2.334),npt));
T3d=zeros(length(N3d));
for i=1:length(N3d) tic();Laplace3d(N3d[i],N3d[i],N3d[i]);T3d[i]=toq();end;
println("      completed!");
```

```julia
using PyPlot
PyPlot.xlabel("taille de la matrice")
PyPlot.ylabel("temps en s")
PyPlot.title("Temps d'assemblage")
plot(N1d,T1d,color="red")
plot(N2d.^2,T2d,color="blue")
plot(N3d.^3,T3d,color="green")
PyPlot.legend(["1d" ; "2d" ;"3d"],loc=2);
```

## En conclusion

A la vue du graphique précédent sur le temps d'assemblage :
* le 1d (courbe rouge) est visiblement linéaire **mais coûte plus chère que le 2d et 3d !!!!**.
* le 2d (courbe bleue) et le 3d (courbe verte) sont clairement linéaires.

**Les courbes obtenues avec surtout l'inversion du 1d laissent à penser que nos méthodes d'assemblage ne sont pas optimales !**. 

Pour rajouter à cela en terme de performance on pourrait :
* Utiliser le fait que la matrice soit symétrique pour n'assembler et ne stocker qu'une partie triangulaire.
* L'usage d'un stockage "sparse" adpaté serait également plus optimal. Par défaut JULIA utilise un format CSC (Compressed Sparse Column), ici un format SKS (Skyline Matrix Storage) serait plus optimal en mémoire et pour le calcul du produit matrice/vecteur.

# 3. Algorithme du Gradient Conjugué préconditionné

## L'algorithme :

* `Precond` est une fonction passée en paramètre, on peut passer la fonction identité avec `x->x` en paramètre i.e. `pcg(A,b,x->x,x0,1e-12,1000)`.
* Le test d'arrêt est calculé sous la forme $\dfrac{\|r\|}{\|b\|} \leq \varepsilon$ i.e. le résidu normalisé par le second membre.

```julia
function pcg(A,b,Precond,x0,Eps,itermax)
    # Renormalisation et test
    bnrm=norm(b);
    if bnrm==0
        return zeros(b)
    end
    #initialisation
    x=x0;
    r=b-A*x;
    z=Precond(r);
    d=copy(z);
    res=[norm(r)/bnrm];
    tmp=dot(r,z);
    iter=1;
    
    while (res[iter]>Eps) & (iter<itermax)
        Ad=A*d;
        alpha=tmp/dot(d,Ad);
        x=x+alpha*d;
        r=r-alpha*Ad;
        z=Precond(r);
        tmp1=dot(r,z);
        beta=tmp1/tmp;
        d=z+beta*d;
        tmp=tmp1;
        iter=iter+1;
        append!(res,[norm(r)/bnrm]);
    end
    return x,res
end
```

## Préconditionnement - préambule

En premier lieu on définit 2 fonctions de résolution (efficace) de sytèmes triangulaires (inf et sup). Ces 2 fonctions remplacent aventageusement celles utilisées par la fonction `\` de JULIA. Les matrices creuses sont en format CSC.

```julia
function triusolve(A,b)
    if !istriu(A) throw(DimensionMismatch("A is'nt upper triangular matrix")) end
    n = length(b)
    if isa(b, Vector)
        nrowB = n
        ncolB = 1
    else
        nrowB, ncolB = size(b)
    end
    ncol = sqrt(prod(size(A)))
    if nrowB != ncol throw(DimensionMismatch("A is $(ncol)X$(ncol) and b has length $(n)")) end
   
    x=zeros(n);
    for i=n:-1:1
        x[i]=(b[i]-x[i])/A[i,i];
        for j=A.colptr[i]:A.colptr[i+1]-2;
            x[A.rowval[j]]+=A.nzval[j]*x[i]
        end
    end
    return x
end
```

```julia
function trilsolve(A,b)
    if !istril(A) throw(DimensionMismatch("A is'nt lower triangular matrix")) end
    n = length(b)
    if isa(b, Vector)
        nrowB = n
        ncolB = 1
    else
        nrowB, ncolB = size(b)
    end
    ncol = sqrt(prod(size(A)))
    if nrowB != ncol throw(DimensionMismatch("A is $(ncol)X$(ncol) and b has length $(n)")) end
   
    x=zeros(n);
    for i=1:n
        x[i]=(b[i]-x[i])/A[i,i];
        for j=A.colptr[i]+1:A.colptr[i+1]-1;
            x[A.rowval[j]]+=A.nzval[j]*x[i]
        end
    end
    return x
end
```

## Préconditionnement - Diagonal

On utilise une variable global $D$ afin de stocker la diagonale de la matrice.

Le calcul se fait en 2 étapes, dans un premier temps on assemble le préconditionneur (ici $D$) puis à chaque itération on procède à la résolution du système donné par le préconditionneur. Le choix en terme de coût et d'inverser 1 fois la diagonale de la matrice et ensuite de multiplier à chaque itération (la division coûte plus chère que la multiplication). 


```julia
function init_PrecDiag(A)
    global D
    D=1 ./ diag(A);
    return "PrecDiag initialised"
end

function PrecDiag(r)
    global D
    z= D .* r;
end
```

```julia
function PrecDiag(r)
    global D
    z= D .* r;
end
```

## Préconditionnement - SSOR

On utilise les variable global $U$, $L$, $D$ et $w$ afin de stocker la décomposition de la matrice de préconditionnement :
$$ M_\omega= \dfrac{\omega}{2-\omega}\left(\frac{D}{\omega}+E\right)\,D^{-1}\,\left(\frac{D}{\omega}+E^T\right) \quad E\text{ partie strictement supérieure de }A$$

Le calcul se fait en 2 étapes, dans un premier temps on assemble le préconditionneur :
* $L \leftarrow \left(\frac{D}{\omega}+E\right)$
* $D \leftarrow \frac{2-\omega}{\omega} D^{-1}$
* $U\leftarrow\left(\frac{D}{\omega}+E^T\right)$.

A chaque itération on procède à la résolution du système à l'aide des 2 fonctions précédentes de résolution de système triangulaires.


```julia
function init_SSOR(A,w_in)
    global U,L,D,w
    w=copy(w_in)
    D=diag(A)/w;
    U=triu(A,1)+spdiagm(D);
    L=tril(A,-1)+spdiagm(D);
    D=D/(2-w);
    return "SSOR initialised"
end

function PrecSSOR(r)
    global U,L,D,w
    r=trilsolve(L,r);
    r=D.*r
    r=triusolve(U,r);
end
```

## Evolution du résidu

Regardons pour une matrice de taille respectable (Laplace1d, 2d et 3d) l'évolution du résidu (normalisé) par $b$, regardons l'effet des préconditionneurs sur la décroissance du résidu nous utiliserons la valeur optimal pour SSOr : $w_{opt}=\frac{2}{1+\sin(\pi h)}$.

### Résidu - Laplace1d

```julia
nn=1000000;
A=Laplace1d(nn);
b=ones(nn);
h=1/(nn+1);
wopt=2/(1+sin(pi*h));

## Sans préconditionnement
(x,res1)=pcg(A,b,x->x,zeros(b),1e-12,400);
## préconditionnement diagonal
init_PrecDiag(A)
(x,res2)=pcg(A,b,PrecDiag,zeros(b),1e-12,400);
## Préconditionnement SSOR
init_SSOR(A,wopt)
(x,res3)=pcg(A,b,PrecSSOR,zeros(b),1e-18,400);

PyPlot.xlabel("Itération k")
PyPlot.ylabel("Norme du résidu")
PyPlot.title("Evolution du résidu - Laplace 1d de taille $nn")
semilogy(1:length(res1),res1,color="red")
semilogy(1:length(res2),res2,color="blue")
semilogy(1:length(res3),res3,color="green")
PyPlot.legend(["Sans préconditionneur" ; "Précond. Diagonal" ;"Précond. SSOR"],loc=4)
grid("on")
```

### Constation

* Les itérations sans préconditionnement semblent ne pas converger ou très très doucement...
* Le préconditionnement diagonal est sans effet (par rapport au non préconditionnement). Cela est normal car la diagonale est de la forme une constante fois l'identité, on ne change pas le conditionnement d'une matrice en a multipliant par une constante !
* Le péconditionnement SSOR est efficace, seulement 18 itérations pour converger vers un résidu en $10^{-16}$ pour une matrice de taille $10^6$. 

### Résidu Laplace2d

```julia
nn=1000;
A=Laplace2d(nn,nn);
b=ones(nn^2);
h=1/(nn+1);
wopt=2/(1+sin(pi*h));

## Sans préconditionnement
(x,res1)=pcg(A,b,x->x,zeros(b),1e-12,400);
## préconditionnement diagonal
init_PrecDiag(A)
(x,res2)=pcg(A,b,PrecDiag,zeros(b),1e-16,400);
## Préconditionnement SSOR
init_SSOR(A,wopt)
(x,res3)=pcg(A,b,PrecSSOR,zeros(b),1e-18,400);

PyPlot.xlabel("Itération k")
PyPlot.ylabel("Norme du résidu")
PyPlot.title("Evolution du résidu - Laplace 2d de taille $(nn*nn)")
semilogy(1:length(res1),res1,color="red")
semilogy(1:length(res2),res2,color="blue")
semilogy(1:length(res3),res3,color="green")
PyPlot.legend(["Sans préconditionneur" ; "Précond. Diagonal" ;"Précond. SSOR"],loc=5)
grid("on")
```

### Constation

* La convergence sans préconditionnement est plus marquée qu'en 1d mais pas extra-ordinaire !
* Le préconditionnement diagonal est toujours sans effet...
* Le péconditionnement SSOR est toujours aussi efficace, avec seulement 18 itérations pour converger vers un résidu en $10^{-16}$ pour une matrice de taille $10^6$.

# 4. Conclusions

Nous venons de voir que la bibliothèque UMFPACK possède des limites vites atteintes (particulierement en 3d). Le gradient conjugué avec un préconditionnement SSOR est une alternative, elle permet au moins dans le cas 3d d'atteindre des taille de matrices plus importantes, le côut mémoire étant fortement réduit. Le préconditionnement SSOR réduit fortement le nombre d'itérations du Gradient conjugué.

**Mais le gradient conjugué préconditionné SSOR est-il vraiment plus performant que la bibliothèque UMFPACK ?**

Il faut donc comparer les temps de résolution, en prenant également en compte le temps de l'assemblage du préconditionneur !

```julia
npt=12
println("1d... ");
N1d=iround(logspace(2,7,npt));
T1d1=zeros(length(N1d));
for i=1:length(N1d)
    A=Laplace1d(N1d[i]); b=ones(N1d[i]);
    wopt=2/(1+sin(pi/(N1d[i]+1)));
    println("  n=$(N1d[i])")
    tic();
    init_SSOR(A,wopt)
    (x,rr)=pcg(A,b,PrecSSOR,zeros(b),1e-16*N1d[i],400);
    T1d1[i]=toq();
end;
println("           completed!");

println("2d... ");
N2d=iround(logspace(1,3.5001,npt));
T2d1=zeros(length(N2d));
for i=1:length(N2d)
    A=Laplace2d(N2d[i],N2d[i]);b=ones(N2d[i].^2)
    wopt=2/(1+sin(pi/(N1d[i]+1)));
    println("  n=$(N2d[i])")
    tic();
    init_SSOR(A,wopt)
    (x,rr)=pcg(A,b,PrecSSOR,zeros(b),1e-16*N2d[i]^2,400);
    T2d1[i]=toq();
end;
println("          completed!");

println("3d...");
N3d1=iround(logspace(1,2.3334,npt)); 
T3d1=zeros(length(N3d1));
for i=1:length(N3d1)
    A=Laplace3d(N3d1[i],N3d1[i],N3d1[i]);b=ones(N3d1[i].^3)
    wopt=2/(1+sin(pi/(N1d[i]+1)));
    println("  n=$(N3d1[i])")
    tic();
    init_SSOR(A,wopt)
    (x,rr)=pcg(A,b,PrecSSOR,zeros(b),1e-16*N3d1[i].^3,400);
    T3d1[i]=toq();
end;
println("          completed!");
```

```julia
PyPlot.xlabel("taille de la matrice")
PyPlot.ylabel("temps en s")
PyPlot.title("Temps de résolution UMFPACK - GC SSOR")
loglog(N1d,T1d,color="red")
loglog(N1d,T1d1,color="red",linestyle="--")
loglog(N2d.^2,T2d,color="blue")
loglog(N2d.^2,T2d1,color="blue",linestyle="--")
loglog(N3d.^3,T3d,color="green")
loglog(N3d1.^3,T3d1,color="green",linestyle="--")
PyPlot.legend(["1d umfpack"; "1d pgc" ; "2d umfpack" ;"2d pgc" ;"3d umfpack" ;"3d pgc"],loc=2)
grid("on")
```

### Analyse et conclusion

Regardons ces dernières courbes :
* On voit que UMFPACK côute cher à l'allumage, il est peu compétitf sur de petites tailles de matrice...

Pour des matrices de grande taille regardons la comparaison UMFPACK/GC-SSOR :
* En 1d le GC est très compétitif cela est du au très faible nombre d'itération pour converger - avantage UMFPACK - Il est possible d'implémenter un pivot de Gauss spécifique pour une matrice tridiagonale ([Algorithme TDMA](http://www.cfd-online.com/Wiki/Tridiagonal_matrix_algorithm_-_TDMA_(Thomas_algorithm)). 
* Le 2d UMFPACK garde l'avantage, le GC met trop d'itérations... Attention on ne compte pas le côut mémoire...
* Pour le 3d cette fois la comparaison est en faveur du GC ! Et ceci à double titre : le GC fait mieux ou aussi bien à toutes les tailles de matrice et en plus il permet d'atteindre des dimensions non-envisageable avec UMFPACK.

Le Gradient Conjugué préconditionné SSOR est une méthode facile d'implémentation. Sont efficacité à $w_{opt}=\frac{2}{1+\sin(\pi h)}$ le rend compétitif en 3d. Quand serait-il avec un préconditionneur ILU, SPAI ou Multigrille...
